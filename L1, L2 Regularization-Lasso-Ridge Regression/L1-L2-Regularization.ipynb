```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[cols_to_use]
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset,drop_first=True).astype(int)
dataset.head()
```

```python
X = dataset.drop('Price',axis=1)
Y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.score(x_train,y_train)
```

<h3>L1(Lasso) Regularization </h3>

```python
from sklearn import linear_model
lesso_reg = linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
```

```python
lesso_reg.fit(x_train,y_train)
```

```python
lesso_reg.score(x_test,y_test)
```

```python
lesso_reg.score(x_train,y_train)
```

<h3>L2(Ridge) Regularization </h3>

```python
from sklearn import linear_model
ridge_reg = linear_model.Ridge(alpha=50,max_iter=100,tol=0.1)
```

```python
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
ridge_reg.score(x_train,y_train)
```

```python
import joblib
joblib.dump(lesso_reg,'L1 Regularization Model')
joblib.dump(ridge_reg,'L2 Regularization Model')
```
